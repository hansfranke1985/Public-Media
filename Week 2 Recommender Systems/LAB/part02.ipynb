```python
# import libraries
import pandas as pd

# load the dataset
books_df = pd.read_csv('data/books/books.csv', low_memory=False)

# quick preview
len(books_df["user_id"].unique())
```

```python
books_df[books_df['user_id'] == 11400].head(4)
```

```python
from itertools import permutations

# create a function that makes combinations of the books that a user (x) reviewed
def create_combinations(x):
    combinations = pd.DataFrame(list(permutations(x.values, 2)), columns=['book_a', 'book_b'])
    return combinations
```

```python
avg_user = books_df.groupby(['user_id']).size().reset_index(name='counts')
avg_user = avg_user.set_index('user_id')
avg_user
```

```python
avg_user = avg_user[avg_user['counts'] > 2]
users_index = avg_user.index

smaller_selection = books_df.iloc[users_index]
smaller_selection
```

```python
smaller_selection.shape
```

```python
# use the create_combinations function

book_combinations = smaller_selection.groupby('user_id')['book_title'].apply(create_combinations)

book_combinations.head(4)
```

```python
# calculate the occurences of book a and book b

book_combinations = book_combinations.reset_index(drop=True)
combi_count = book_combinations.groupby(['book_a', 'book_b']).size()
combi_count = combi_count.to_frame(name='size').reset_index()
combi_count.sort_values('size', ascending=False).head(5)
```

```python
# search for a specific
combi_count[combi_count['book_a'] == 'Life of Pi'].sort_values('size', ascending=False).head(10)
```
