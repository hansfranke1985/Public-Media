```python
#define a function to average ratings 
def weighted_rating(x, m, C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C)

def filter_weights(df_agg, vote_avg, vote_count, threshold):
    # C is the mean vote or rating across the whole dataframe
    C = df_agg[vote_avg].mean()
    print(C)

    #m is the minimum votes or ratings to be listed
    m = df_agg[vote_count].quantile(threshold)
    print(m)

    q_df = df_agg.copy().loc[df_agg[vote_count] > m]
    
    return q_df

```
