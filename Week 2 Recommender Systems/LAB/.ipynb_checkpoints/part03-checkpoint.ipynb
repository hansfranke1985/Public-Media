```python
# import libraries
import pandas as pd

# load the dataset
books_df = pd.read_csv('data/books/books.csv', low_memory=False)

# quick preview
books_df.head(5)
```

```python
# remove duplicates
books_df2 = books_df.drop_duplicates(subset=['book_title'])

# create smaller selection
books_df2 = books_df2[(books_df2['Summary'].str.contains('Hobbit')) & (books_df2['Category'] != '9')]

books_df2.head(5)
```

```python
# get rid of those pesky []'s and '
books_df2['Category'] = books_df2.Category.apply(lambda x: x.replace('[', ''))
books_df2['Category'] = books_df2.Category.apply(lambda x: x.replace(']', ''))
books_df2['Category'] = books_df2.Category.apply(lambda x: x.replace("'", ""))

books_df2['Category2'] = books_df2['Category'] + ', ' + books_df2['Language']

# cast to list
books_df2['Category2'] = books_df2.Category2.str.split(',').tolist()
books_df2['Category2'].head(10)
```

```python
# explode into rows
books_df2 = books_df2.explode('Category2')
books_df2.head(5)
```

```python
# let's make a crosstab
books_category = pd.crosstab(books_df2['book_title'], books_df2['Category2'])
```

```python
books_category
```

```python
# import necessary stuff
from sklearn.metrics import jaccard_score
from scipy.spatial.distance import pdist, squareform

# calculate the distances
jaccard_distances = pdist(books_category.values, metric='jaccard')
square_jaccard_distances = squareform(jaccard_distances)

square_jaccard_distances
```

```python
jaccard_similarity_array = 1 - square_jaccard_distances
jaccard_similarity_array
```

```python
distance_df = pd.DataFrame(jaccard_similarity_array, index=books_category.index, columns=books_category.index)
distance_df
```

```python
distance_df['The Tolkien Reader'].sort_values(ascending=False)
```
