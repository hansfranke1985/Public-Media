```python
# import libraries
import pandas as pd

# load the dataset
steam_df = pd.read_csv('data/steam/steam_games.csv', low_memory=False)

# quick preview
steam_df.head(5)
```

```python
steam_df.info()
```

```python
#just look for my favorite game ever :)
steam_df['name'] = steam_df['name'].astype('str')
steam_df2 = steam_df[( steam_df['name'].str.contains('Dota 2') )]
steam_df2.head()
```

```python
#reduce to a small dataset (description of games with strategy)

#convert to str
steam_df['popular_tags'] = steam_df['popular_tags'].astype('str')

steam_df = steam_df[(steam_df['popular_tags'].str.contains('Strategy'))]

steam_df
```

```python
# Transform GENRE to LIST

#convert to str
steam_df['genre'] = steam_df['genre'].astype('str')

# cast to list
steam_df['genre2'] = steam_df.genre.str.split(',').tolist()
steam_df['genre2']
```

```python
# explode into rows
steam_df2 = steam_df.explode('genre2')
steam_df2.head(5)
```

```python
# let's make a crosstab
games_category = pd.crosstab(steam_df2['name'], steam_df2['genre2'])
games_category
```

```python
# import necessary stuff
from sklearn.metrics import jaccard_score
from scipy.spatial.distance import pdist, squareform

# calculate the distances
jaccard_distances = pdist(games_category.values, metric='jaccard')
square_jaccard_distances = squareform(jaccard_distances)

square_jaccard_distances
```

```python
jaccard_similarity_array = 1 - square_jaccard_distances
jaccard_similarity_array
```

```python
#Create a DF to present in a better visual
distance_df = pd.DataFrame(jaccard_similarity_array, index=games_category.index, columns=games_category.index)
distance_df
```

```python
distance_df['Dota 2'].sort_values(ascending=False)
```

```python
#filter only > 90% similarity
distance_df['Dota 2'][distance_df['Dota 2'] > 0.9].sort_values(ascending=False)
```
