```python
# import libraries
import pandas as pd

# load the dataset
movies_metadata = pd.read_csv('data/movies/movies_metadata.csv', low_memory=False)
ratings = pd.read_csv('data/movies/ratings.csv', low_memory=False)

```

```python
#convert id to int64
#remove bad indexes (shit!!)
movies_metadata = movies_metadata.drop([19730, 29503, 35587])
movies_metadata['id'] = movies_metadata['id'].astype("int")

#Combine the datasets

merged_movies = movies_metadata.merge(ratings, left_on='id', right_on='movieId')
merged_movies.head()
```

```python
from itertools import permutations

# create a function that makes combinations of the books that a user (x) reviewed
def create_combinations(x):
    combinations = pd.DataFrame(list(permutations(x.values, 2)), columns=['item_a', 'item_b'])
    return combinations
```

```python
#look for user reviews
avg_user = merged_movies.groupby(['userId']).size().reset_index(name='counts')
avg_user = avg_user.set_index('userId')
avg_user.reset_index().sort_values('counts', ascending=False)
```

```python
#Smaller selection
avg_user = avg_user[avg_user['counts'] > 200]
users_index = avg_user.index

smaller_selection = merged_movies.iloc[users_index]
smaller_selection.head()
```

```python
# use the create_combinations function

movie_combinations = smaller_selection.groupby('userId')['original_title'].apply(create_combinations)

movie_combinations.head(4)
```

```python
# calculate the occurences of book a and book b

movie_combinations = movie_combinations.reset_index(drop=True)
combi_count = movie_combinations.groupby(['item_a', 'item_b']).size()
combi_count = combi_count.to_frame(name='size').reset_index()
combi_count.sort_values('size', ascending=False).head(5)
```

```python
# search for a specific
combi_count[combi_count['item_a'] == 'Star Wars'].sort_values('size', ascending=False).head(10)
```
